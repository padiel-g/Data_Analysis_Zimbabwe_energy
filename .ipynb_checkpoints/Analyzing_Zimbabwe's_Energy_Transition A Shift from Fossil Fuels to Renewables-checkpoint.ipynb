# Data Preparation

```python
import pandas as pd
import numpy as py
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\cyphe\Downloads\energy-and-mining_zwe.csv")
```

```python
df.info()
```

```python
print('Dataset Shape', df.shape)
```

```python
df.isna().sum
```

```python
# Drop the second row which contains metadata. using .iloc[1:] to select all rows from the second one onwards
df = df.iloc[1:].copy()

# Convert `Year` and `Value` columns to numeric types for plotting and calculations
df['Year'] = pd.to_numeric(df['Year'])
df['Value'] = pd.to_numeric(df['Value'])

# Display the cleaned DataFrame information to confirm data types are correct
print("Cleaned Data Info:")
print(df.info())

# Display the unique indicator names to see what data is available for analysis
print("\nUnique Indicator Names:")
print(df['Indicator Name'].unique())
```

# Analyzing Fossil Fuels vs. Renewables

```python
# Filter for the relevant indicators for fossil fuels and renewables
fossil_renewables_df = df[df['Indicator Name'].isin([
    'Fossil fuel energy consumption (% of total)',
    'Renewable energy consumption (% of total final energy consumption)'
])].copy()

# Make indicator names more concise for the plot legend
fossil_renewables_df['Indicator Name'] = fossil_renewables_df['Indicator Name'].replace({
    'Fossil fuel energy consumption (% of total)': 'Fossil Fuels',
    'Renewable energy consumption (% of total final energy consumption)': 'Renewables'
})

# Create the line plot
plt.figure(figsize=(12, 7))
for indicator, data in fossil_renewables_df.groupby('Indicator Name'):
    plt.plot(data['Year'], data['Value'], label=indicator, marker='o')

plt.title('Fossil Fuels vs. Renewables Consumption in Zimbabwe (1990-2021)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage of Total Energy Consumption (%)', fontsize=12)
plt.legend(title='Energy Source', fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Comparing Hydropower to Other Renewables

```python
# Filter for the relevant indicators: Hydropower and other renewables
hydro_renewables_df = df[df['Indicator Name'].isin([
    'Electricity production from hydroelectric sources (% of total)',
    'Electricity production from renewable sources, excluding hydroelectric (% of total)'
])].copy()

# Make indicator names more concise for the plot legend
hydro_renewables_df['Indicator Name'] = hydro_renewables_df['Indicator Name'].replace({
    'Electricity production from hydroelectric sources (% of total)': 'Hydropower',
    'Electricity production from renewable sources, excluding hydroelectric (% of total)': 'Other Renewables (excluding Hydro)'
})

# Create the line plot
plt.figure(figsize=(12, 7))
for indicator, data in hydro_renewables_df.groupby('Indicator Name'):
    plt.plot(data['Year'], data['Value'], label=indicator, marker='o')

plt.title('Hydropower vs. Other Renewables in Zimbabwe (1990-2021)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage of Total Electricity Production (%)', fontsize=12)
plt.legend(title='Energy Source', fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Exploring Correlation with Economic Indicators

```python
# Prepare data for the third plot: Correlation with economic indicators.
# using `GDP per unit of energy use (PPP $ per kg of oil equivalent)`
gdp_renewables_df = df[df['Indicator Name'].isin([
    'Renewable electricity output (% of total electricity output)',
    'GDP per unit of energy use (PPP $ per kg of oil equivalent)'
])].copy()

# Pivot the table to have years as index and indicators as columns
gdp_renewables_pivot = gdp_renewables_df.pivot_table(
    index='Year',
    columns='Indicator Name',
    values='Value'
).reset_index()

# Rename columns for clarity
gdp_renewables_pivot.rename(columns={
    'Renewable electricity output (% of total electricity output)': 'Renewable_Output',
    'GDP per unit of energy use (PPP $ per kg of oil equivalent)': 'GDP_Per_Energy_Use'
}, inplace=True)

# Create the scatter plot
plt.figure(figsize=(12, 7))
plt.scatter(
    gdp_renewables_pivot['GDP_Per_Energy_Use'],
    gdp_renewables_pivot['Renewable_Output'],
    s=100,
    alpha=0.7,
    edgecolors='k'
)
plt.title('Renewable Energy vs. Economic Indicator in Zimbabwe (1990-2021)', fontsize=16)
plt.xlabel('GDP per Unit of Energy Use (PPP $ per kg of oil equivalent)', fontsize=12)
plt.ylabel('Renewable Electricity Output (% of total)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
